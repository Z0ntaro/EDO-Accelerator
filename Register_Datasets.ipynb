```python
from notebookutils import mssparkutils
import pandas as pd

files = mssparkutils.fs.ls("Files")

datasets = []

for f in files:
    if f.name.lower().endswith(".xlsx"):
        datasets.append({
            "dataset_name": f.name.replace(".xlsx",""),
            "category": "Auto",
            "active": True
        })

df = pd.DataFrame(datasets)

spark.sql("DROP TABLE IF EXISTS dataset_registry")

spark.createDataFrame(df).write.mode("overwrite").saveAsTable("dataset_registry")
display(df)
```
